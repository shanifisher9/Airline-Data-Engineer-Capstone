#### The Data gives more information about airline codes. I got the data from [here](https://www.census.gov/foreign-trade/reference/codes/aircarrier/acname.txt)

```python
# Reading in the data. Since it's a text file I needed a few more parameters while adding it in.
airline_codes = pd.read_csv('Airline Codes.txt', sep='\t', engine='python', names=["Carrier Name", "IATA", "ICAO"])
# Since the data is all 1 long row I had to take the last 3 charachters and designate them to ICAO
airline_codes['ICAO'] = airline_codes['Carrier Name'].str.strip().str[-3:]
# Needed to do the same thing as above so got all the needed information for the IATA column
airline_codes["IATA"]=airline_codes["Carrier Name"].str[-21:-10]
# Removing any whitespaces in the column. Used this resource for help: https://stackoverflow.com/questions/41476150/removing-space-from-dataframe-columns-in-pandas
airline_codes["IATA"]=airline_codes["IATA"].str.replace(' ','')
# Did the same thing for the first column. I wanted to remove anything that I didn't need and was part of the ICAO and IATA columns
airline_codes["Carrier Name"]=airline_codes["Carrier Name"].str[:-22]
# Stripping the white spaces from the of the columns
airline_codes["Carrier Name"]=airline_codes["Carrier Name"].str.lstrip()
airline_codes["ICAO"]=airline_codes["ICAO"].str.lstrip()
# Removing the first line since it repeats the column names
airline_codes = airline_codes.iloc[1:]
# removing the last 5 lines since they aren't airlines and caused problem when inserting the data into Redshift
airline_codes = airline_codes.iloc[:-5]
# Checking for duplicates
print(len(airline_codes)-len(airline_codes.drop_duplicates()))
# Looking at shape of the data
print(airline_codes.shape)
# Looking for missing values
print(airline_codes.isnull().sum())
# Filling missing values with "None"
airline_codes= airline_codes.fillna('None')
# Sampling 3 rows of the data
airline_codes.sample(3)
```

## Data Dictonary

### Fact Table: Airport
##### Airport: Airport Name
##### Region: Country And State the airport is located in ex: US-TX - United States, Texas
##### Country: Country the airport is located in ex: US - United States
##### Airport_IATA: A Three-letter Geocode defined by the International Air Transport Association ex: ORD - Chicago O'Hare International Airport

### Dimension Table: Airport_Flights_2020
##### Airline Primary KEY: Airline Name 
##### Airport_IATA_Destination:A Three-letter Geocode defined by the International Air Transport Association Of Where the flight is leaving
##### Airport_IATA_Arriving:A Three-letter Geocode defined by the International Air Transport Association Of Where the flight is arriving
##### Airline_IATA: IATA codes, are two-character codes assigned by the International Air Transport Association ex: AA - American Airlines
##### Distance: How far the flight is travelling
##### Passengers: How many passengers on the flight
##### Year: Year of the flight
##### Month: Month of the flight

### Dimension Table: Airlines
##### Airline_IATA Primary Key: IATA codes, are two-character codes assigned by the International Air Transport Association ex: AA - American Airlines
##### Airline: Name of Airline
##### ICAO: The International Civil Aviation Organization (ICAO) gave each airline a specific code to go by
##### Country: Country The Airline is From
##### Active: If the Airline is currently active

### Dimension Table: Past_Flights:
##### Airport_destination PRIMARY KEY: 3 Letter abbrevation(IATA) which airport the flight is leaving from
##### Airport_arriving:3 Letter abbrevation(IATA) which airport the flight is arriving to
##### Date: Date of the flight
##### Seats: Amount of seats on the flight
##### Distance: Distance of the flight

### Dimension Table: Airport_Description
##### Airport PRIMARY KEY: Airport Name
##### Airport_IATA: A Three-letter Geocode defined by the International Air Transport Association
##### Latitude:Latitude of the airport
##### Longitude: Longidute of the airport
##### City: City of the airport
##### Region: Country And State the airport is located in ex: US-TX - United States, Texas
##### Country: Country of the airport
##### Continent: Continent of the airport
##### Airport_type: What type of airport ex: Small, Large, Helipad
##### Elevation: What elevation is the airport
##### GPS_code: The GPS code of the airport

### Dimension Table: Countries
##### Country PRIMARY KEY: Country Name
##### Country_abbreviation: Country name abbreviated ex: US - United States
##### Continet: Continent of the country
##### Wikipedia_link: Wikipedia Link in case you want more information about the country

### Dimension Table: Population
##### Airport PRIMARY KEY: Three-letter Geocode defined by the International Air Transport Association
##### City: City of the population
##### Population: Amount of people in the city

### Inserting Data Into Redshift Tables

##### I had a problem with data types so used this [link](https://stackoverflow.com/questions/21500509/casting-text-to-date-in-redshift) to fix the issue.

```python
airports_insert_table = ("""INSERT INTO airports(Airport, Region, City, Country, Airport_IATA)
                        SELECT DISTINCT ad.name,
                        ad.iso_region,
                        ad.municipality,
                        ad.iso_country,
                        ad.iata_code
                        FROM airports_with_description AS ad;
""")
airlines_insert_table = ("""INSERT INTO airlines(Airline_IATA, Airline_name, ICAO, Country, Active)
                        SELECT DISTINCT ac.IATA,
                        ar.name,
                        ac.icao,
                        ar.country,
                        ar.active
                        FROM airline_codes AS ac JOIN airlines_and_countries AS ar on ar.IATA=ac.IATA AND ar.icao=ac.icao;
""")
airport_description_insert_table = ("""INSERT INTO airport_description(Airport, Airport_IATA, Latitude, Longitude, City, Region, Country, Continent, Airport_type,  Elevation, GPS_Code)
                        SELECT DISTINCT name,
                        iata_code,
                        CAST(latitude_deg as float) latitude,
                        CAST(longitude_deg as float) longitude,
                        municipality,
                        iso_region,
                        iso_country,
                        continent,
                        type,
                        elevation_ft,
                        gps_code
                        FROM airports_with_description;
""")
country_insert_table = ("""INSERT INTO country(Country, Country_abbreviation, Continent, Wikipedia_link)
                        SELECT DISTINCT country.name,
                        country.code,
                        country.continent,
                        country.wikipedia_link
                        FROM airport_countries AS country;
""")
airport_flights_2020_insert_table = ("""INSERT INTO airport_flights_2020(Airline, Airport_IATA_destination, Airport_IATA_arriving, Airline_IATA, Distance, Passengers, Year, Month)
                        SELECT DISTINCT f.unique_carrier_name,
                        f.dest,
                        f.origin,
                        f.unique_carrier,
                        distance,
                        f.passengers,
                        f.year,
                        f.month
                        FROM flights_2020 AS f;
""")
population_insert_table = ("""INSERT INTO population(Airport, City, Population)
                        SELECT DISTINCT a_past.origin_airport,
                        ad.municipality,
                        cast(a_past.Origin_population as int)
                        FROM airports_past_flights AS a_past JOIN airports_with_description AS ad ON ad.iata_code=a_past.origin_airport;
""")
past_flights_insert_table = ("""INSERT INTO past_flights(Airport_destination, Airport_arriving, Date, Passengers, Seats, Distance)
                        SELECT DISTINCT a_past.destination_airport,
                        a_past.origin_airport,
                        CASE WHEN a_past.fly_date SIMILAR TO '[0-9]{4}-[0-9]{2}-[0-9]{2}'
                        OR a_past.fly_date SIMILAR TO '[0-9]{4}-[0-9]{2}-[0-9]{2} [0-9]{2}:[0-9]{2}:[0-9]{2}' 
                        THEN TO_DATE(a_past.fly_date, 'YYYY-MM-DD')
                        ELSE TO_DATE('1900-01-01', 'YYYY-MM-DD') END as Date
                        CONVERT(date, a_past.fly_date) as date,
                        a_past.passengers,
                        CAST(a_past.seats as int) as seats,
                        CAST(a_past.distance as int) as distance
                        FROM airports_past_flights AS a_past;
""")
```

##### Taking all variable names so they could be looped through

```python
copy_tables_from_S3toRedshift = [airports_with_description_copying, airport_countries_copying, airlines_and_countries_copying, airports_past_flights_copying, airline_codes_copying, flights_2020_copying]
creating_tables = [airports_creating_table, airlines_creating_table, airport_description_creating_table,country_creating_table ,airport_flights_2020_creating_table, population_creating_table, past_flights_creating_table,airports_with_description_creating_table,airport_countries_creating_table,airlines_and_countries_creating_table,airports_past_flights_creating_table,airline_codes_creating_table,flights_2020_creating_table]
dropping_tables_if_exist = [airports_drop_table, airlines_drop_table, airport_description_drop_table,country_drop_table ,airport_flights_2020_drop_table, population_drop_table, past_flights_drop_table, airports_with_description_drop_table,airport_countries_drop_table,airlines_and_countries_drop_table,airports_past_flights_drop_table,airline_codes_drop_table,flights_2020_drop_table]
insert_data_in_tables = [airports_insert_table, airlines_insert_table, airport_description_insert_table, country_insert_table,airport_flights_2020_insert_table, population_insert_table, past_flights_insert_table]
```

##### 1. Looping Through All the Variable to Drop Existing Tables
##### 2. Create All Tables
##### 3. Copying Data From S3 to Redshift
##### 4. Insert Data Into the Tables

```python
def dropping_the_tables(cur, conn):
    
    """Dropping tables if exist """
    
    for query in dropping_tables_if_exist:
        cur.execute(query)
        conn.commit()


def creating_the_tables(cur, conn):
    
    """Creating Tables"""
    
    for query in creating_tables:
        cur.execute(query)
        conn.commit()


def copying_data_from_s3(cur, conn): 
    
    """Copying Data From S3 Bucket"""
    
    for query in copy_tables_from_S3toRedshift:
        cur.execute(query)
        conn.commit()


def inserting_data_into_tables(cur, conn):
    
    """Inserting Data Into Tables From S3 Bucket"""
    
    for query in insert_data_in_tables:
        cur.execute(query)
        conn.commit()
```

#####  Running The Functions Create above

```python
dropping_the_tables(cur, conn)
creating_the_tables(cur, conn)
copying_data_from_s3(cur, conn)
inserting_data_into_tables(cur, conn)
```
