```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
bike_share=pd.read_csv('bike-share.csv')
```

```python
bike_share.sample(10)
```

```python
ax = sns.countplot(x="day_of_week", data=bike_share)
```

```python
ax = sns.barplot(x="day_of_week", y="count", data=bike_share)
```

```python
g = sns.catplot(x="day_of_week", y="count", col="workingday", hue='holiday',data=bike_share, kind="swarm",  height=4, aspect=.7);
```

```python
sns.jointplot("humidity", "count", data=bike_share, kind='reg');
sns.jointplot("windspeed", "count", data=bike_share, kind='reg');
sns.jointplot("temp", "count", data=bike_share, kind='reg');
```

```python
plt.title("Random Walk Example", fontsize=16, fontweight='bold')
plt.suptitle("Random Walk Suptitle", fontsize=10)
plt.xlabel("Number of Steps")
plt.ylabel("Accumulative Sum")
plt.show()
```
