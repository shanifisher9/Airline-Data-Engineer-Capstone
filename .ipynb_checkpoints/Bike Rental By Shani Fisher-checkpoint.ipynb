### Importing Data

Imported packages that I need for analysis such as pandas to read the dataframe and seaborn to make visualizations.

Read data to get a better understanding of the columns and features.

Did not have any trouble/problems with the data, and decided not to do any feature engineering.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
```

```python
bike_share=pd.read_csv('bike-share.csv')
bike_share.sample(10)
```

### Bike Rentals On Weekday and Weekend

To see this data clearly I decided to make a bar plot. On the X axis is days of the week and on the Y axis is the number of bike rentals.

This bar plot shows the amount of bikes rented on each day. There does not seem to be a drastic difference between the bike rentals on the weekday and the weekend. Additionally it seems like Thursday and Friday are the most popular days for bike shares.

In the second plot it shows the weekdays and weekends in different graphs. This illustrates the same point but is a different way of looking at the data. Additionally, the days that are not working days and have bike rentals are in fact holidays.

```python
rentals_by_day_of_week = sns.barplot(x="day_of_week", y="count", data=bike_share)
plt.title("Days Of The Week", fontsize=16, fontweight='bold')
plt.xlabel("Days of Week")
plt.ylabel("Number of Rentals")
plt.show()
```

```python
rentals_by_weekday_and_workday = sns.catplot(x="day_of_week", y="count", col="workingday", hue='holiday',
                data=bike_share, kind="swarm",  height=4, aspect=.7);
```

### Temperature And Bike Rentals

In this plot we can see the actual temperature outside and the amount of bikes rented in each weather condition. The temperature is on the X axis and number of rentals is on the Y axis.

There is definitly a correlation between the temperature and amount of bike rentals. I would say that from 25-35 C is the most common temperature to bike.

Additionally I made a graph with the column temp as the X axis. However, I decided to use the column atemp instead because the results were similar and the acutal temperature would matter more to the bike renters than the forcasted temperature.

```python
atemp_on_rentals = sns.lineplot(x="atemp", y="count", data=bike_share)
plt.suptitle("Temperature Effect on Bike Rentals", fontsize=14)
plt.xlabel("Temperature")
plt.ylabel("Number of Rentals")
plt.show()
```

### Non-Linear Dependancy

Lastly, the graph shows the humidity on the X axis and the number of bike rentals on Y axis.

I found that there is really no linear pattern between humidity and the number of bike renters. Additionally, I looked at the columns temp, atemp and wind and those all seem to have some linear dependancy. However, there was not as much linear dependancy with wind as there was with temp.

```python
sns.jointplot("humidity", "count", data=bike_share, kind='reg');
plt.xlabel("Humidity")
plt.ylabel("Number of Rentals")
plt.show()
```
